```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

fig, ax = plt.subplots(figsize=(10,10))

ax.plot(x, y)

ax.set(title="Exemplo de plot Customize", xlabel="Ax X", ylabel="Ax Y")
plt.show()  
```

```python
produtos = {"Arroz": 1.80,
            "Farinha": 1.30,
            "Massa": 0.98}

fig, ax=plt.subplots()
fig, bx=plt.subplots()
ax.barh(produtos.keys(), produtos.values())
bx.bar(produtos.keys(), produtos.values())

ax.set(title="Lista de Produtos", xlabel="produtos", ylabel="preo ")
```
