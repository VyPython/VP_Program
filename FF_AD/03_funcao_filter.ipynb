A função filter() do pandas é usada para filtrar linhas de um DataFrame baseado em uma condição específica. Ela retorna um novo DataFrame com as linhas que satisfazem a condição dada.

```python
import pandas as pd
df = pd.read_csv('sales_data.csv')
#df.info()
# filtrar as linhas onde a coluna "Customer_Age" é maior que 50
#age_maior_50 = df.filter(['Customer_Age'] > 50) tem de ser usado de forma diferente
'''exemplo
idades = [25, 30, 55, 60, 45]
maiores_50 = list(filter(lambda idade: idade > 50, idades))
print(maiores_50) 
# Saída: [55, 60]'''
#age_maior_50 = (df['Customer_Age'] > 50 & (df['Countr#y'] == 'United States'))

age_maior_50 = df[df['Customer_Age'] > 50]

#age_maior_50 = (df['Customer_Age'] > 50) & (df['Country'] == 'United States')
age_maior_50
```

A função filter() também pode ser usada para filtrar dados com base em uma função personalizada. Por exemplo:

```python
import pandas as pd
df = pd.read_csv('sales_data.csv')

data_range = (df['Date'] >= '2013-01-01') & (df['Date'] <= '2013-12-31')
data_range
print(df[data_range])
```
